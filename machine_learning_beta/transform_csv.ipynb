```python
import pandas as pd 

def string_list_to_float_list(entry):
    entry = entry.strip("[]\n,").split(' ')
    new_list = []
    for word in entry:
        word = word.strip(',')
        if (word == '[')or(word == '')or(word == ']')or(word == ','):
            continue
        else:
            new_list.append(float(word))
    return new_list

def transform_embeddings (df):
    '''
    transform a string list to a float list
    '''

    rows = len(df)
    matrix = []
    column_names = []
    
    for row in range(rows):
        matrix.append(string_list_to_float_list(df.iloc[row].embeddings))

    for col in range(len(matrix[0])):
        column_names.append('embbedings_'+str(col))
    
    embeddings_df = pd.DataFrame(matrix,columns=column_names)

    return embeddings_df

def transform_csv(filename:str,save_csv:bool):
    '''
    breaks down the embeddings into different columns
    '''

    df = pd.read_csv(filename)
    df_embeddings = transform_embeddings (df)
    df_cat = pd.concat([df,df_embeddings],axis=1)
    df_cat=df_cat.drop(['embeddings'],axis=1)
    display(df_cat.head(3))
    
    if save_csv:
        newfilename = filename.split('.')[0]+'_transformed.csv'
        df_cat.to_csv(newfilename)
    
    

```
